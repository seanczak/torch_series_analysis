### Inference to the Rescue

As noted above, the answer we're looking for is: what is the composition of the marbles in the bag?  In other words, what percentage/proportion of the marbles are blue and what percentage/proportion are red? Since Biden wins Illinois, let's call the porportion of blue marbles "$p$" which allows us to say that if $p$>50% (also written as $p$ > 0.5), then Biden wins.  We've already established that we're going to use a sample of $n$ marbles to estimate this $p$. That value will be whatever the proportion of blue marbles we observe in our drawing (naturally).  Since this value (called an estimator) is not necessarily the true value $p$ inside the bag (because we could have gotten unlucky with our sample and be wrong), let's call it something else that's similar but not the same. In stats we give $p$ a cute little hat like this, $\hat{p}$ to say: "this is not actually $p$ but it's our best guess." Ok so to recap

- $p$: actual proportion of marbles that are blue
- $\hat{p}$: our observed proportion of blue marbles in a sample drawn from the bag (serves as our best guess at p)
- $n$: the number of marbles drawn from the bag

And we can rephrase our problem in a simplified manner to start to tackle in systematically.

Goals:
- Determine values our marble bag's $p$ could be. (Note: if $p$ > 50% blue then Biden wins, else Trump wins)

Constraints
- Do it with as few counted marbles, $n$, as possible
- Don't be wrong

If only there was a way to say, I'm 99.99% confident that $p$ is between these two numbers.  I'm glad you asked because that is exactly what we invented confidence intervals for.  They give us a way to take some observe data and use it to infer something about the world that produced that data.  In our case, we want to know a range that we believe the marble bag's proportion of blue marbles $p$ will fall into (remember if we know $p$ we know who won). So how do we do this. Well, that's where statistical inference comes in. 

In order to be accessible, I plan to be brief about this and then we'll jump back. {deep inhaling}

It turns out that the marble drawing situation is called a hypergeometric process.  For those of you who don't care what that is, just know that the important part is that it has a name which means that someone smart figured out what to do with it. In fact, that was how I simulated all the plots so far (python will just do it for you if you tell it the name of the distribution). Going deeper it turns out that when the number of marbles in the bag increases it approaches a different (easier to work with) situation called a binomial process. The only reason I bring any of this up is that the Binomial process has one important parameter and it's called, you guessed it $p$ which we can easily approximate a confidence interval for.  I don't plan on going into the approach that approximates the confidence interval on $p$ but I'll say that if you're interested look up some of these terms: binomial distribution, confidence interval, Bayesian inference, conjugate prior. 

{big exhale}

For the rest of us, let's just look at pictures shall we? Let's say that we were those humans from before that had drawn 49/100 red balls from a bag that contains only God knows how many blue balls (again, $p$ = proportion blue balls).  So our $\hat{p}$ estimated from this is.. you guessed it 0.49 (that's how statisticians write 49%). Given that n=100, what can we say about $p$?  Well, using the aforementioned, not to be mentioned again, Bayesian inference approach, we can generate a guess as what our distribution of $p$ is when n=100 and 49 of them are blue.  

```python
p_hat = 0.49
nsample = 1e2
m = Experiment(p_hat,nsample)
axes = m.plot_beta()
axes[0].axvline(0.5, c='r',linewidth=4)
```

Again, the red line is at the decision threshold where if $p$ is to the left - Trump wins and if it's to the right - Biden wins.  We see that there is a decent probability that $p$ could be anywhere from 0.35 to 0.65. Of course, the majority of the probability is centered around $p$ being 0.49 (which we know to be wrong). However, due to the large "spread" in possible values for $p$, we still don't know who will win. Let's increase n = 1000 and say that the results now show something more realistic like 540 blue balls (i.e. $\hat{p}$ ~ 0.54) which is closer to bag's true proportion of blue balls ($p$ = 0.55).

```python
p_hat = 0.55
nsample = 1e3
m = Experiment(p_hat,nsample)
axes = m.plot_beta(xlims=[0.3,0.7])
axes[0].axvline(0.5, c='r',linewidth=4)
```

When we do this, the spread of the distribution has greatly decreased (i.e. the distribution has gotten thinner).  If we increased $n$ again, now to 10,000 marbles counted, we can see that the spread continues to shrink to almost nothing.  In fact, it shrinks so much that we see that the probability that $p$ is less than 0.5 (Trump winning) shrinks with it. This means that even though we haven't counted all the samples, we have a very narrow window with which to guess $p$ and none of them spell victory for Trump. In fact, the magenta lines I've drawn on all the plots shown the boundaries which hold 99.99% of the probability between them. This means we have a 99.99% confidence that p is between these two values. So by the time we get to this point in the counting process. We are already 99.99% confident that $p$ is between 0.53 and 0.57.  In other words, even before counting 10,000 marbles, we knew that Biden would win.

```python
p_hat = 0.55
nsample = 1e4
m = Experiment(p_hat,nsample)
axes = m.plot_beta(xlims=[0.3,0.7])
axes[0].axvline(0.5, c='r',linewidth=4)
for bound in [0.53,0.57]: axes[0].axvline(bound, c='m',linewidth=1)
```

If we were to use the one bag model above, we'd make the eggregious error of assuming that we could draw from any of these bags and make a general statement about the entire state.  Well why not? Well, what if instead of assuming a single bag of votes, we assumed that each county was a smaller bag within that bag. Yes, the state's marble bag is actually a bag of county marble bags (the rabbit hole continues!).  

Ok with this assumption, we could count a small number of votes from one of those county marble bags and determine a confidence interval (our range of guess values from above) on $p$ for that specific county.  This is basically a high and low esimate on what $p$ could be (lets call them $p_high$ and $p_low$).  We could then multiply each of these bounds by the expected total votes in that county to get an approximation of the range of votes that each county produces.  If we did this for all the county's, we would have a solid estimate of what to expect the election result would be.

```python
import json
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
plt.style.use('ggplot')
plt.style.use(['dark_background'])
```

```python
with open('virginia.json') as json_file:
    data = json.load(json_file)

virginia_counties = []
cols = ['name', 'uid','princincts_reporting', 'n_princincts', 'updated', 'Biden', 'Trump', 'percent_Biden','percent_Trump']

for locality in data['Localities']:
    name = locality['Locality']['LocalityName']
    uid  = int(locality['Locality']['LocalityCode'])
    reporting = locality['PrecinctsReporting']
    nPrecincts = locality['PrecinctsParticipating']
    updated = locality['LastModified']
    # results of the candidates
    results = pd.DataFrame.from_dict(locality['Candidates'])
    biden = results.query('BallotName == "Joseph R. Biden"')
    trump = results.query('BallotName =="Donald J. Trump"')
    nBiden = biden.Votes.values[0]
    nTrump = trump.Votes.values[0]
    perBiden = biden.Percentage.values[0]
    perTrump = trump.Percentage.values[0]
    county_results = (name, uid, reporting, nPrecincts, updated, nBiden, nTrump, perBiden, perTrump)
    virginia_counties.append(county_results)
vir = pd.DataFrame(virginia_counties, columns= cols)
vir['total'] = vir.Biden + vir.Trump
for percent_column in ['percent_Biden','percent_Trump']:
    vir[percent_column] = vir[percent_column].str.rstrip('%').astype('float') / 100.0
vir.head()
```

```python
df = vir.sort_values('total',ascending=False).iloc[:30]

x = 'name'
y = 'votes'
hue = 'candidate'

melted_df = pd.melt(df[['Biden','Trump', x]], id_vars=[x],var_name=hue, value_name = y)

font = {'family' : 'normal',
        'size'   : 14}
matplotlib.rc('font', **font)

plt.figure(figsize=(20,6))
sns.barplot(x=x, y=y, hue=hue, data=melted_df, palette=['m','y'])
plt.xticks(rotation=90); plt.xlabel('')
```

```python
vir['Biden10k'] = (vir.percent_Biden * 1e4).astype(int)
vir['Trump10k'] = (vir.percent_Trump * 1e4).astype(int)
vir.loc[vir.total < 1e4, 'Biden10k'] = vir.loc[vir.total < 1e4, 'Biden']
vir.loc[vir.total < 1e4, 'Trump10k'] = vir.loc[vir.total < 1e4, 'Trump']
vir.head()
```

Calculate the CI

```python
def calc_CI(p1, N, eps = 1e-15):
    # z score given an alpha of 0.0001 (two tailed)
#     z = 1.96
#     z = 2.576 #99%
    z = 3.891 # 99.99
    # normal approximation using CLT
    interval = z * np.sqrt(p1 * (1 - p1) / N)
    # upper lower bounds are just +/- above
    upper = p1 + interval
    lower = p1 - interval
    # bound them to (0,1) - this made vectorization break, it worked anyways
#     upper = 1 - eps if upper > 1 else upper
#     lower = 0 + eps if lower < 0 else lower
    return upper, lower

# what is p_hat that biden wins for each county
vir['p_hat10k'] = vir.Biden10k / (vir.Trump10k + vir.Biden10k)
# calculate the CI
N = vir.Biden10k+vir.Trump10k
p = vir.p_hat10k
bounds = calc_CI(p, N)
bounds = pd.DataFrame(bounds,['p_high','p_low']).T
vir2 = pd.merge(vir,bounds,right_index=True,left_index=True)
vir2.head()
```

Make the predictions

```python
# calculate raw numbers of votes predicted
vir2['predict_high'] = (vir2.p_high * vir2.total).astype(int)
vir2['predict_low'] = (vir2.p_low * vir2.total).astype(int)

# grab them after summing over the rows
high, low, total = vir2.sum()[['predict_high','predict_low', 'total']]

# combine into a dataframe after calculating the p as well
results = [(low, high, total),
           (f'{np.round(low/total*100,2)}%',f'{np.round(high/total*100,2)}%','NA')]
results = pd.DataFrame(results,index = ['Votes','Percent'],columns = ['Low Estimate','High Estimate','Total Virginia Votes'])
results
```

beta function, find alpha and beta as a function n and p https://www.real-statistics.com/binomial-and-related-distributions/beta-distribution/

```python
class Experiment():
    def __init__(self, p_hat, nsample, N=1e6):
        '''
        N is total sample size (if not infinite)
        p is proportion that is "True"
        nsample is the amount to be sampled
        '''
        self.nsample = int(nsample)
        self.p       = p_hat
        # for normal approx
        self.mean    = self.nsample * p_hat
        self.var     = self.nsample * p_hat * (1 - p_hat)
        self.std     = np.sqrt(self.var)
        # for hypergeometric
        self.K       = p_hat*N
        self.N       = N
        # for beta
        self.a       = self.mean
        self.b       = self.nsample - self.mean
        return
    
    def plot_binomial(self,N = int(1e6), bins = None, xlims = None, ylims = None):
        
        fig, axes = plt.subplots(1,2,figsize = (20,5))

        # plot pmf
        ax = axes[0];
        # generate a random sample and plot it
        x = np.random.binomial(self.nsample,self.p,N)
        axes[0].hist(x, bins = 50,color='grey',edgecolor='w')
        # set x lims 
        xlims = [self.mean - 5*self.std , self.mean + 5*self.std] if xlims is None else xlims
        ax.set_xlim(xlims)
        ylims = [0,N//10] if ylims is None else ylims
        ax.set_ylim(ylims)
        # plt pmf
        x = np.arange(0,nsample,step = 1)
        cdf = scipy.stats.binom.cdf(x,self.nsample,self.p)
        axes[1].plot(x,cdf, c= 'c')
        axes[1].set_title('cdf')
        return axes
        
    def plot_hypergeometric(self, N = int(1e6), bins = None, xlims = None, ylims = None):
        
        fig, axes = plt.subplots(1,2,figsize = (20,5))

        # plot pmf
        ax = axes[0];
        # generate a random sample and plot it
        ngood = self.K
        nbad  = self.N - ngood
        x = np.random.hypergeometric(ngood, nbad, self.nsample, size=N)
        axes[0].hist(x, bins = 50, color='grey', edgecolor='w')
        # set x lims 
        xlims = [self.mean - 5*self.std , self.mean + 5*self.std] if xlims is None else xlims
        ax.set_xlim(xlims)
        ylims = [0,N//10] if ylims is None else ylims
        ax.set_ylim(ylims)
        # plt pmf
        x = np.arange(0,nsample,step = 1)
        cdf = scipy.stats.hypergeom.cdf(x,self.N, ngood, self.nsample)
        axes[1].plot(x,cdf, c= 'c')
        axes[1].set_title('cdf')
        return axes
        
    def plot_normal_approx_bin(self,N = int(1e6), bins = None, xlims = None, ylims = None):
        
        fig, axes = plt.subplots(1,2,figsize = (20,5))

        # plot pmf
        ax = axes[0];
        # generate a random sample and plot it
        x = np.random.normal(self.mean, self.std, size=N)
        axes[0].hist(x, bins = 50, color='grey', edgecolor='w')
        # set x lims 
        xlims = [self.mean - 5*self.std , self.mean + 5*self.std] if xlims is None else xlims
        ax.set_xlim(xlims)
        ylims = [0,N//10] if ylims is None else ylims
        ax.set_ylim(ylims)
        # plt pmf
        x = np.arange(0,nsample,step = 1)
        cdf = scipy.stats.norm.cdf(x,self.mean, self.std)
        axes[1].plot(x,cdf, c= 'c')
        axes[1].set_title('cdf')
        return axes
        
    def plot_beta(self,N = int(1e6), bins = None, xlims = None, ylims = None):
        
        fig, axes = plt.subplots(1,2,figsize = (20,5))

        # plot pmf
        ax = axes[0];
        # generate a random sample and plot it
        x = np.random.beta(self.a, self.b, size=N)
        axes[0].hist(x, bins = 50, color='grey', edgecolor='w')
        # set x lims 
        xlims = [(self.mean - 5*self.std)/self.nsample , (self.mean + 5*self.std)/self.nsample] if xlims is None else xlims
        ax.set_xlim(xlims)
        ylims = [0,N//10] if ylims is None else ylims
        ax.set_ylim(ylims)
        # plt pmf
        x = np.linspace(0,1,N)
        cdf = scipy.stats.beta.cdf(x,self.a, self.b)
        axes[1].plot(x,cdf, c= 'c')
        axes[1].set_title('cdf')
        return axes
    
    def plot_normal_approx_beta(self,N = int(1e6), bins = None, xlims = None, ylims = None):
        
        fig, axes = plt.subplots(1,2,figsize = (20,5))

        # plot pmf
        ax = axes[0];
        # generate a random sample and plot it
        x = np.random.normal(self.p, self.std/self.nsample, size=N)
        axes[0].hist(x, bins = 50, color='grey', edgecolor='w')
        # set x lims 
        xlims = [(self.mean - 5*self.std)/self.nsample , (self.mean + 5*self.std)/self.nsample] if xlims is None else xlims
        ax.set_xlim(xlims)
        ylims = [0,N//10] if ylims is None else ylims
        ax.set_ylim(ylims)
        # plt pmf
        x = np.linspace(0,1,N)
        cdf = scipy.stats.norm.cdf(x,self.p, self.std/self.nsample)
        axes[1].plot(x,cdf, c= 'c')
        axes[1].set_title('cdf')
        return axes
```

Visually inspecting validity of normal approx for binomial and hypergeometric
