## Plotting imports and functions

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
plt.style.use(['dark_background'])

def plot_hist(x, bins = 100, ax = None, title = None, xlabel = '', ylabel = '', c='c'):
    '''might want to change this to seaborn'''
    if ax is None:
        plt.figure()
        ax = plt.gca()
    ax.hist(x, bins=bins, color = c)
    ax.set_xlabel(xlabel); ax.set_ylabel(ylabel)
    ax.set_title(title)
    return ax

def plot_scatter(x, y, ax = None, c = 'c', s = 5, alpha = 1,
                title = None, xlabel = '', ylabel = ''):
    if ax is None:
        plt.figure()
        ax = plt.gca()
    ax.scatter(x, y, c=c, s=s, alpha=alpha)
    ax.set_xlabel(xlabel); ax.set_ylabel(ylabel)
    ax.set_title(title)
    return ax
```

# Create Fake Data
of form y = ax + b + noise. Also give a helper method to plot the data next to a histogrm of the noise generated for it

## Class

```python
class LinData():
    '''Creates data in the form `y = ax + b + noise`
    Args
        - a,b [float]: parameters of lin data
        - N [int]: length of x vector
        - noise_func [function]: function to create noise given only x
    '''
    def __init__(self, a, b, N, noise_func, noise_args = [], xlims = [0,1]):
        self.N = N
        # shape x to have two dims (for sklearn)
        self.x = np.linspace(xlims[0], xlims[1], N).reshape(N,1)
        # generate a vector for noise in the shape of x
        self.noise = noise_func(self.x)
        # create y_clean and then add noise to it
        self.y_clean = a * self.x + b
        self.y = self.y_clean + self.noise
        return
    
    def plot_raw(self, title = None, bins = 100):
        # create multiple plots
        fig, axs = plt.subplots(1, 2, figsize = (20,4))
        fig.suptitle(title, fontsize = 16)
        # y vs x
        plot_scatter(self.x, self.y, ax = axs[0], c = 'c', s = 5, alpha = 0.5, xlabel='x', ylabel = 'y')
        # noise hist
        plot_hist(self.noise, bins=bins, ax = axs[1], xlabel='y_noise', ylabel = 'Frequency')
        return axs
```

## No noise data

```python
def no_noise_gen(x): return np.zeros(x.shape)

clean_data = LinData(a = 10, b = 3, N = 10000, noise_func = no_noise_gen)
clean_data.plot_raw('No Noise Data')
```

## Gaussian Noise Data

```python
def gaussian_noise_gen(x, mu = 0, sigma = 0.5): 
    return np.random.normal(mu, sigma, size = x.shape)

norm_data = LinData(a = 10, b = 3, N = 10000, noise_func = gaussian_noise_gen)
axes = norm_data.plot_raw('Gaussian Data')
# add the underlying system line
axes[0].plot(norm_data.x,norm_data.y_clean, c='r')
axes[0].legend(['original y','data'])
```

## Increasing Gaussian Noise Data
The noise increases with x. Sigma(x) = x (multiplying normal dist by scalar: https://math.stackexchange.com/questions/275648/multiplication-of-a-random-variable-with-constant and product of multiplying two normal dist https://www.johndcook.com/blog/2012/10/29/product-of-normal-pdfs/)

```python
def incr_gaussian_noise_gen(x, mu = 0, sigma = 1): 
    return np.random.normal(mu, sigma, size = x.shape) * x

inc_norm_data = LinData(a = 10, b = 3, N = 10000, noise_func = incr_gaussian_noise_gen)
axes = inc_norm_data.plot_raw(title = 'Gaussian Data', bins = 1000)
# add the underlying system line
axes[0].plot(inc_norm_data.x,inc_norm_data.y_clean, c='r')
axes[0].legend(['original y','data'])
```

## Log Normal Noise Data
We can see that now the data "sits above" the mean (of the underlying normal distribution)

```python
def lognorm_noise_gen(x, mu = 0, sigma = 0.5): 
    '''Note that the mean and standard deviation are not the values 
    for the distribution itself, but of the underlying normal distribution 
    it is derived from'''
    return np.random.lognormal(mu, sigma, size = x.shape)

lognorm_data = LinData(a = 10, b = 3, N = 10000, noise_func = lognorm_noise_gen)
axes = lognorm_data.plot_raw('Log Normal Data')
# add the underlying system line
axes[0].plot(lognorm_data.x,lognorm_data.y_clean, c='r')
axes[0].legend(['original y','data'])
```

## LinReg Class
Inherits from sklearn's Ridge but simply fits immediately then stores the estimate of the variances at each x given the models' predictions

```python
from sklearn.linear_model import Ridge

def rmse(preds, targs): return np.sqrt(((preds - targs) ** 2).mean())

class LinReg(Ridge):
    def __init__(self,x,y, alpha = 1):
        LR.__init__(self, alpha = alpha)
        self.x, self.y = x, y
        self.fit(x,y)
        # store the predictions and squared residuals
        self.preds = self.predict(x)
        self.resids2 = (self.preds - self.y) ** 2
        return
    
    def rmse_score(self, x, y):
        preds = self.predict(x)
        targs = y
        return rmse(preds, targs)
```

## No noise data model
We can see that the model rmse is super low and is exactly the same the approximation for stdev. This is good because it should be 0 and only isn't because we've added a reg turm
